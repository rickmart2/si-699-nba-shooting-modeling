```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
feature_data = pd.read_csv('feature_data.csv')
cas_target_data = pd.read_csv('cas_target_data.csv')
test_players = pd.read_csv('test_players.csv')['0']
```

```python
cas_data = pd.merge(feature_data, cas_target_data, left_on='Name', right_on='Player', how='inner')
cas_data = cas_data.drop(columns=['Player'])
```

## Test-Train Split

```python
test_df = cas_data[cas_data['Name'].isin(test_players)]
train_df = cas_data[~cas_data['Name'].isin(test_players)]
```

```python
X_train = train_df.drop(columns=['Name', 'NBA_C&S_PPS'])
y_train = train_df['NBA_C&S_PPS']

X_test = test_df.drop(columns=['Name', 'NBA_C&S_PPS'])
y_test = test_df['NBA_C&S_PPS']
```
