```python
full_otd_attempts_sorted = full_otd_attempts.sort_values(by="FG Att", ascending=False)
eligible_otd = full_otd_attempts_sorted[full_otd_attempts_sorted['FG Att'] > 50]
eligible_otd["NBA_OTD_PPS"] = (2 * eligible_otd["2 FG Made"] + 3 * eligible_otd["3 FG Made"]) / eligible_otd["FG Att"]
otd_target_data = eligible_otd[["NBA_OTD_PPS"]]
```
