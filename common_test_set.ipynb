```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
feature_data = pd.read_csv('feature_data.csv')
cas_target_data = pd.read_csv('cas_target_data.csv')
otd_target_data = pd.read_csv('otd_target_data.csv')
```

```python
cas_data = pd.merge(feature_data, cas_target_data, left_on='Name', right_on='Player', how='inner')
cas_data = cas_data.drop(columns=['Player'])
otd_data = pd.merge(feature_data, otd_target_data, left_on='Name', right_on='Player', how='inner')
otd_data = otd_data.drop(columns=['Player'])
```

```python
common_players = list(set(cas_data['Name']).intersection(set(otd_data['Name'])))
_, test_players = train_test_split(common_players, test_size=0.15, random_state=42)
```

```python
pd.Series(test_players).to_csv("test_players.csv", index=False)
```
